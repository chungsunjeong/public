```python
def str_cleaner(s):
    return s.replace('\'','').replace('\n','').replace('[','').replace(']','').replace(')','').replace('(','').replace(' ','')

from xml.etree.ElementTree import parse
import pandas as pd
seperator='\t'
```

```python
drugbank_file='D:\Wisdom\workspace_python/research\dataset\db_drugbank\drug-target_drugbank_preprocessed.xml'.replace('\\','/')
drugbank_file
xml_tree=parse(drugbank_file)
```

```python
root=xml_tree.getroot()
drugs=root.findall("drug")
dict_drug=set()
dict_target=set()
for drug in drugs:
        # Several constraints
        name = drug.find("name")
        targets_element = drug.find("targets")
        if targets_element is not None:
            targets = targets_element.findall("target")
            for target in targets:
                actions = target.find("actions")
                action = actions.findall("action")
                action_list=[]
                for act in action:
                    action_list.append(act.text)

                polypeptide = target.find("polypeptide")
                drug_id = drug.find("drugbank-id").text
                generic_name = name.text.lower()
                if polypeptide is None:
                    target_name = target.findtext("name")
                    target_id = target.findtext("id")
                    dict_drug.update(("\t".join([drug_id,generic_name])))
                    dict_target.update(("\t".join([target_id,target_name])))
                else:
                    target_name = target.findtext("name")
                    target_id = polypeptide.attrib["id"]
                    dict_drug.update(["\t".join([drug_id,generic_name])])
                    dict_target.update(["\t".join([target_id,target_name])])

# f_w=open(preprocessed_file,'w')
# description = '\t'.join(['drug ID', 'target UniprotID',
#                           'drug name','target name', 'actions'])
# f_w.write(description + '\n')
# for pair in l_drug_target:
#     f_w.write(pair+'\n')
#     n_rule+=1
# print('Total number drug-target rule from drugbank: '+str(n_rule))
```

```python
f=open('drug_name.txt','w')
for a in dict_drug:
    try:
        f.write(a+'\n')
    except:pass
f.close()
```
