```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
```

```python
x=tf.Variable(2.)
y=tf.Variable(x.initialized_value()+3.)
print(x)
init=tf.global_variables_initializer()
sess= tf.Session()
sess.run(init)
print(sess.run(x))
print(sess.run(y))
sess.close()
```

```python
x=tf.Variable(tf.random_normal([784,100],stddev=0.35))
y=tf.Variable(x.initialized_value()+3.)
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
print(sess.run(x))
print(sess.run(y))
print(y.get_shape())
```

```python
saver=tf.train.Saver()
# saver=tf.train.Saver(max_to_keep=4, keep_checkpoint_every_n_hours=2)
# 최근 2시간동안 4개의 모델만 저장한다.
save_path='./tmp/test.ckpt'
saver.save(sess,save_path)
# saver.save(sess, save_path, global_step=1000) 
# 1000 반복 마다 model 저장

# saver.save(sess, save_path, global_step=step,write_meta_graph=False) 
# 모델 값만 저장하고 graph 는 저장안함
sess.close()
```

```python
x=tf.Variable(tf.random_normal([784,100],stddev=0.35))
y=tf.Variable(x.initialized_value()+3.)
init=tf.global_variables_initializer()
saver=tf.train.Saver()

sess=tf.Session()
sess.run(init)
saver.restore(sess,'./tmp/variable.ckpt')
print(sess.run(x))
print(sess.run(y))
print(y.get_shape())
```
